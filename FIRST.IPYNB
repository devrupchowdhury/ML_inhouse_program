The .flatten() function is used to convert a multi-dimensional NumPy array into a one-dimensional array.

Returns a new 1D array

Original array remains unchanged

Default order is row-major (C order)

```python
# Converting 2D array to 1D array
import numpy as np  
array_2d = np.array([[1, 2, 3], [4, 5, 6]])  
array_1d = array_2d.flatten()   
print("1D Array:", array_1d)
```

Array Slicing in NumPy
Array slicing is used to extract a part (sub-array) from a NumPy array using index ranges
array[start : stop : step]


```python
import numpy as np
a = np.array([10, 20, 30, 40, 50])
print("Original Array:", a)
```

```python
print(a[1:4] )    # [20 30 40]
print(a[:3])      # [10 20 30]
print(a[2:])      # [30 40 50]
print(a[::2])     # [10 30 50]
print(a[::-1])    # Reverse array
```

Normalization of an Array (Easy Explanation)

Normalization is the process of scaling array values to a defined range so that all values are on a similar scale.
It is mainly used in Machine Learning, Data Science, and Statistics to improve model performance.

```python
import numpy as np
# Create a NumPy array
array = np.array([5,10,15,20,25])
Array_max=arr.max()
Array_min = arr.min()
Arr = (array-Array_max)/(Array_max-Array_min)
print("Normalized Array:", Arr)
```

The diagonal() function extracts diagonal elements along specified axes. For a 3D array, setting axis1=1 and axis2=2 extracts the main diagonal from each 2D slice.

```python
#Retrieve diagonal from an array
import numpy as np
arr_3d = np.arange(27).reshape(3,3,3)
print("3D Array is:\n",arr_3d)
print("DIAGONAL ELEMENTS FOR AXIS1=1 AND AXIS2 ARE:")
print(arr_3d.diagonal(axis1=1 , axis2=2))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
subjects = ['Math', 'Science', 'History', 'Art', 'Physical Education']
scores = [85, 90, 78, 92, 88]
y_pos = np.arange(len(subjects))
plt.barh(y_pos, scores, color='cyan', edgecolor='black' , height=0.5 )
plt.xlabel('Scores')
plt.ylabel('Subjects')
plt.title('Student Scores by Subject')
plt.show()
#if i want to change the width of the bars then we can use plt.barh() and specify the width parameter like plt.barh(y_pos, scores, height=0.5)
```
